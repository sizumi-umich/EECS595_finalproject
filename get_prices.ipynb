```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime
import yfinance as yf
```

```python
ticker = pd.read_csv('ticker.csv',index_col=0)
```

```python
results = pd.DataFrame()
for symbol in ticker["Symbol"].values:
    try:
        stock = yf.Ticker(symbol)
        prices = stock.history(start="2010-01-01")
        close = pd.DataFrame({symbol:prices["Close"].values},index=prices.index.strftime('%Y/%m/%d').values)
        results = pd.concat([results, close], axis=1, join='outer')
        print("successfully downloaded for {}".format(symbol))
    except:
        print("failed to process for {}".format(symbol))
```

```python
results.to_csv("sp500.csv")
```
